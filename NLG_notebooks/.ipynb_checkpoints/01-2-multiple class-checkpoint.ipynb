```python
# load amt
amt_hGI_pos = load_pickle('data/amt_hGI_pos.pickle')

# add amt tags
for i, v in dic.items():
    if i in amt_hGI_pos.keys():
        v['AMT'] = amt_hGI_pos[i] 
ls = list(amt_hGI_pos.keys())
```

```python
world_ls = [i for i, v in dic.items() if 'World Cuisine' in v['sections']]
print('%d recipes are included in World cuisine' %(len(world_ls)))
world_related = []
for i, v in dic.items():
    world_related += sum([cat for cat in dic[i]['sec']['categories'] if 'world cuisine' in cat],[])
world_related = list(set(world_related))
print('%d categories are included under world cuisine'%(len(world_related)-1))

# store the categorical info to another dic, called dic2
dic2 = {}
for i, v in dic.items():
    dic2[i] = {'sec' : v['sec']}
for cat in world_related:
    for i, v in dic2.items():
        tags = v['sec']['tags']
        dic2[i][cat] = 1 if cat in tags else 0
# then turn dic2 to df
df = pd.DataFrame.from_dict(dic2, orient = 'index')

summ = df.sum(axis = 0)
summ1000 = summ[summ>1000]
print('%d categories has higher than 1000 recipes' % (len(summ1000)))
print('%s the following categories has higher than 1000 recipes' % (summ1000.index))
df_15 = df[summ1000.index]
```

```python
health_ls = [i for i, v in dic.items() if 'Diet and Health' in v['sections']]
print('%d recipes are included in diet and health'%(len(health_ls)))
```

```python
diet_related = []
for i, v in dic.items():
    diet_related += sum([cat for cat in dic[i]['sec']['categories'] if 'diet and health' in cat],[])
diet_related = list(set(diet_related))
print('%d categories are included under diet and health'%(len(diet_related)-1))
```

```python
# store the categorical info to another dic, called dic2
dic2 = {}
for i, v in dic.items():
    dic2[i] = {'sec' : v['sec']}
for cat in diet_related:
    for i, v in dic2.items():
        tags = v['sec']['tags']
        dic2[i][cat] = 1 if cat in tags else 0
# then turn dic2 to df
df = pd.DataFrame.from_dict(dic2, orient = 'index')
```

```python
summ = df.sum(axis = 0)
summ1000 = summ[summ>1000]
print('%d categories has higher than 1000 recipes' % (len(summ1000)))
print('%s the following categories has higher than 1000 recipes' % (summ1000.index))
df_23 = df[summ1000.index]
```

```python
print('show how many recipes in col is covered in row')
print('suggest most recipes are covered by healthy and diet and health')
df_intersect = {}
for col in df_23.columns:
    pos_col = np.nonzero(df_23[col])[0]
    df_intersect[col] = [len(np.intersect1d(pos_col, np.nonzero(df_23[row])[0]))/len(pos_col) for row in df_23.columns]
df_intersect = pd.DataFrame.from_dict(df_intersect)
sns.heatmap(df_intersect, xticklabels = df_intersect.columns, yticklabels = df_intersect.columns)
```
