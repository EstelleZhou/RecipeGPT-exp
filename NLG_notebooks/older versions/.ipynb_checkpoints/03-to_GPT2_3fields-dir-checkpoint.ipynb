```python
%time
%load_ext autotime
%load_ext autoreload
%autoreload 2


# if cannot import the modules, add the parent directory to system path might help

import os
import sys
parent_dir = os.path.abspath(os.getcwd()+'/..')+'/'
sys.path.append(parent_dir)


import os
from utils.path import dir_HugeFiles
from utils.preprocessing import load
from utils.words import make_corpus_0, get_wordcount_list
from utils.save import make_dir, save_pickle, load_pickle, auto_save_csv, print_time
from sklearn.model_selection import train_test_split

import pandas as pd
import numpy as np

#dir_save = os.path.normpath(dir_HugeFiles+'dph/dic_20190607.pickle')
#dic = load(dir_save)
random_seed = 2019
```

```python
dic = load(dir_save = '../big_data/dic_20190830.pickle')

ls = [i for i,v in dic.items() if len(v['ingredients'])>1]
print('drop %d recipes with less than 2 ingredients' %(len(dic)-len(ls)))
ls = [i for i in ls if len(dic[i]['directions'])>1]
print('furthur drop %d recipes with less than 2 instructions' %(len(dic)-len(ls)))
desc = [i for i in ls if len(dic[i]['description'])<1]
print('drop %d recipes with no description' %(len(desc)))
print('now we are using recipe54k %d' % len(ls))
```

```python
def txt(v, fields, mode = 'train'):
    '''
    fields: an order list, the last is the field to predict
    mode: test/train, return string X, y or X+y
    '''
    to_write = []
    for field in fields:
        if field == 'title':
            to_write += ['<start-title>']+v['name']+['<end-title>']
        if field == 'ingredients':
            to_write += ['<start-ingredients>']+[ingr + ' $' for ingr in v[field]]+['<end-ingredients>']
        if field == 'directions':
            to_write += ['<start-directions>']+v[field]+['<end-directions>']
                                                     
    if mode == 'train':
        return ' '.join(to_write)+'\n'
                                                     
    elif mode == 'test':
        field_to_predict = '<start-%s> '%fields[-1]
        to_X, to_y = ' '.join(to_write).split(field_to_predict)
        return to_X+field_to_predict, to_y

class to_gpt2:
    def __init__(self, dic, ls):
        ls_train, self.ls_test, _, __ = train_test_split(ls, ls, test_size = 0.2, random_state = random_seed, shuffle = True)
        self.ls_train, self.ls_val, _, __ = train_test_split(ls_train, ls_train, test_size = 0.25, random_state = random_seed , shuffle = True)
        self.dic = dic
    def train(self, ls, filename, overwrite = False, is_val = False):
        to_write = ''
        for i, v in self.dic.items():
            if i in ls:      
                self.save(filename+'%d'%(i)+'d.txt', txt(v, ['title','ingredients','directions']), overwrite = False)
                self.save(filename+'%d'%(i)+'i.txt', txt(v, ['title','directions','ingredients']), overwrite = False)
                self.save(filename+'%d'%(i)+'t.txt', txt(v, ['ingredients','directions','title']), overwrite = False)
    def test(self, ls, filename, overwrite = False, is_val = False):
        to_write = ''
        for i, v in self.dic.items():
            if i in ls:
                self.save(filename+'X/%d'%(i)+'d.txt', txt(v, ['title','ingredients','directions'], mode = 'test')[0], overwrite = False)
                self.save(filename+'X/%d'%(i)+'i.txt', txt(v, ['title','directions','ingredients'], mode = 'test')[0], overwrite = False)
                self.save(filename+'X/%d'%(i)+'t.txt', txt(v, ['ingredients','directions','title'], mode = 'test')[0], overwrite = False)
                
                self.save(filename+'y/%d'%(i)+'d.txt', txt(v, ['title','ingredients','directions'], mode = 'test')[1], overwrite = False)
                self.save(filename+'y/%d'%(i)+'i.txt', txt(v, ['title','directions','ingredients'], mode = 'test')[1], overwrite = False)
                self.save(filename+'y/%d'%(i)+'t.txt', txt(v, ['ingredients','directions','title'], mode = 'test')[1], overwrite = False)
                
                self.save(filename+'Xy/%d'%(i)+'d.txt', txt(v, ['title','ingredients','directions']), overwrite = False)
                self.save(filename+'Xy/%d'%(i)+'i.txt', txt(v, ['title','directions','ingredients']), overwrite = False)
                self.save(filename+'Xy/%d'%(i)+'t.txt', txt(v, ['ingredients','directions','title']), overwrite = False)
                
    def fast_export(self, filename, overwrite = False):
        self.train(self.ls_train, filename+'train/', overwrite)
        self.train(self.ls_val, filename+'val/', overwrite)
        self.test(self.ls_test, filename+'test/', overwrite)
    def save(self, filename, to_write, overwrite = False):
        make_dir(filename)
        if os.path.isfile(filename) == True and overwrite == False:
            print('already exists'+filename)
        else:    
            with open(filename,'w') as f:
                f.write('%s' % to_write)
```

```python
model = to_gpt2(dic, ls)
#model.fast_export(dir_HugeFiles+'to_gpt2/recipe54k_0827/',overwrite = True)
```
